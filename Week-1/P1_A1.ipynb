```python
import numpy as np
arr=np.loadtxt('https://raw.githubusercontent.com/astroclubiitk/computational-astrophysics/main/Week-1/Assignment/Starfish_Data.csv',delimiter=',')
print(arr)
#array printed for refrence
#column array approach used , 108 rows + 3 columns
```

```python
#defining new arrays to store data
#arrays are 108 rows+1 column
arr1=np.zeros(108).reshape(108,)
arr2=np.zeros(108).reshape(108,)
arr3=np.zeros(108).reshape(108,)

for i in range(108): 
    arr1[i,]=(arr[i,0]/5)+(arr[i,1]/2)+0.034 
    arr2[i,]=10**arr1[i,]                                        #8000 is the total probability 
    arr3[i,]=(arr2[i,]*arr[i,2])/8000 
    
print(arr3[:])
```

```python
#stroing mean distance from of cluster
dist=np.sum(arr3)
print(dist,"pc")
```
