```python
import numpy as np
import matplotlib.pyplot as plt
import astropy
```

```python
from astropy.timeseries import LombScargle
```

```python
#I-band data of a random cephid 
time,I_mag,e_I_mag=np.genfromtxt("http://ogledb.astrouw.edu.pl/~ogle/OCVS/data/I/13/OGLE-LMC-CEP-0013.dat",unpack=True)
```

```python
plt.figure(figsize=(8,8))
plt.scatter(time,I_mag,c='r',s=8)
plt.title("Observed Light Curve")
plt.xlabel("Time(days)")
plt.ylabel("I-band")
plt.show()
```

```python
freq,power=LombScargle(time,I_mag,e_I_mag).autopower(minimum_frequency=0.01,maximum_frequency=10)
plt.plot(freq,power)
plt.xlabel("frequency(day inverse)")
plt.ylabel("Power")
plt.title("P-F spectrum")
plt.show()
```

```python
freq_max=freq[np.argmax(power)]
mp_period=1/freq_max
print (mp_period)
```

```python
phase=np.remainder(time,mp_period)/mp_period
plt.figure(figsize=(8,8))
plt.scatter(phase,I_mag,c='b',s=8)
plt.xlabel("Phase")
plt.ylabel("I-band")
plt.title("OGLE-LMC-CEP-0013",size=18)
plt.show()
```
